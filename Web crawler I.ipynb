```python
#--------------- NOTE: This is a program which is able to scrape max 100 results due to google restrictions policy. It takes very few seconds to give output.............................................


#import libraries---------------------------------------------------------------------------------


import requests
from bs4 import BeautifulSoup
import pandas as pd

#function to scrape data from google search-------------------------------------------------------

def google_search_results(query):
    URL = f"https://www.google.com/search?q={query}&num=10000"
    headers = {
        "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:87.0) Gecko/20100101 Firefox/87.0"
    }
    page = requests.get(URL, headers=headers)
    soup = BeautifulSoup(page.content, "html.parser")
    results = []

    for g in soup.find_all("div", class_=["g Ww4FFb vt6azd tF2Cxc", "g dFd2Tb", "g eejeod"]):
        anchors = g.find_all("a")
        if anchors:
            link = anchors[0]["href"]
            title = g.find("h3").text
            item = {
                "title": title,
                "link": link
            }
            results.append(item)
    return results[:10000]

#print data and write it into csv file------------------------------------------------------

query = "site:youtube.com openinapp.co"
results = google_search_results(query)

df = pd.DataFrame(results)
print(df)
df.to_csv('export_dataframe.csv', index=False, header=True)

# for result in results:
#     print(f"Title: {result['title']}")
#     print(f"Link: {result['link']}")
```
