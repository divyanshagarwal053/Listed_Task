```python
#--------------- NOTE: This is a program I tried to solve problem of google restrictions and getting more than 100 results. It may take more time to give output.............................................


#------------------------------------------import libraries---------------------------------------------------------------
import requests
import time
import random
from bs4 import BeautifulSoup
import pandas as pd

#  ------------------------------------  Function to retrieve/scrap data from google search-------------------------------------------------------------
def crawl(query, page=0):
    url = f"https://www.google.com/search?q={query}&start={page}"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.67 Safari/537.36"
    }
    res = requests.get(url, headers=headers)
    soup = BeautifulSoup(res.text, "html.parser")
    results = []
    for g in soup.find_all("div", class_=["g Ww4FFb vt6azd tF2Cxc", "g dFd2Tb", "g eejeod"]):
        anchors = g.find_all("a")
        if anchors:
            link = anchors[0]["href"]
            title = g.find("h3").text
            item = {"title": title, "link": link}
            results.append(item)
    return results

#---------------------------time delay to avoid google restriction of retriving large data------------------------------

query = "site:youtube.com openinapp.co"
results = []
for i in range(0, 23):
    page = i * 11
    time.sleep(random.uniform(0.5, 1.5))
    page_results = crawl(query, page)
    results.extend(page_results)

#---------------------------- printing the data scraped-----------------------------------------------------------------    
    
print("Total Results:", len(results))
for result in results[:10000]:
    #print(result)
    df = pd.DataFrame(results)
    print(df)

#--------------------------------store data into csv--------------------------------------------------------------------    

df.to_csv('export_dataframe4.csv', index=False, header=True)
```
